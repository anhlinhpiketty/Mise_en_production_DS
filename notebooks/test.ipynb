```python
from dotenv import load_dotenv
import os

BUCKET_NAME = "projet-jocas-prod"
S3_MODEL_PATH = (
    "projet-jocas-prod/Data_etudes/Etude_num/modeles_NER_spacy/camembertav2_skill_only"
)
LOCAL_MODEL_PATH = (
    "/home/onyxia/work/model_spacy_trained"  # O霉 le t茅l茅charger sur le disque
)

SOURCE_PATH = (
    "s3://projet-jocas-prod/JOCAS_V3_NO_DUP_NO_PARTNER_NO_JIT/annee=2025/*/*/*.parquet"
)

S3_ENDPOINT_URL = "https://" + os.environ["AWS_S3_ENDPOINT"]



load_dotenv()
secret = os.environ["SECRET"]
print(secret)
```

```python
import s3fs
import spacy


fs = s3fs.S3FileSystem(
    endpoint_url=S3_ENDPOINT_URL,
    key="jocas-prod",
    secret=secret,
)
if not os.path.exists(LOCAL_MODEL_PATH):
    print("T茅l茅chargement du mod猫le...")
    fs.get(S3_MODEL_PATH, LOCAL_MODEL_PATH, recursive=True)
    print("T茅l茅chargement termin茅.")
else:
    print("Mod猫le d茅j脿 pr茅sent localement, on passe le t茅l茅chargement.")


nlp = spacy.load(LOCAL_MODEL_PATH)
```

```python
text = "Caca oiefozeng霉,a霉zf\nVous devrez avoir des comp茅tences en Machine Learning"
```

```python
with nlp.select_pipes(enable=["transformer", "ner"]):
    nlp_list = nlp.pipe([text])
```

```python
chunk_skill_list = []
for doc in nlp_list:
    # Extraction l茅g猫re (juste des strings)
    ents = [e.text for e in doc.ents]
    chunk_skill_list.append(str(ents))
```
